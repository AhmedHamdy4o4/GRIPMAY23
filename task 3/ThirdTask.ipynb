```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
import plotly.express as px
```

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
df.info()
```

data has no missing values

```python
df.head()
```

```python
df.drop(columns=['Country','Postal Code'],inplace=True)
```

```python
df.describe().T
```

```python
df.select_dtypes('float').columns
```

```python
df.columns
```

```python
sns.relplot(x='Profit',y='Sales',data=df);
```

```python
sns.heatmap(df.corr(),cmap='Blues',annot=True,cbar=False);
```

```python
sns.displot(x='Sales',data=df)
plt.xscale('log')
plt.show();
```

```python
sns.displot(x='Profit',data=df)
plt.xscale('log')
plt.show()
```

```python
sns.set_style('whitegrid')
fig,ax=plt.subplots(figsize=(9,5))
sns.barplot(x='Segment',y='Sales',data=df,ci=False,ax=ax)
plt.xlabel('Segment',fontsize=18)
plt.ylabel('Sales',fontsize=18);
```

```python
df.groupby(['Region'])['Sales'].agg(np.mean).plot(kind='bar')
plt.xlabel('Regions',fontsize=12)
plt.ylabel('Sales',fontsize=12);
```

```python
fig,ax=plt.subplots(figsize=(9,5))
sns.barplot(x='Ship Mode',y='Sales',hue='Region',data=df,ci=False)
plt.ylabel('Sales',fontsize=12)
plt.xlabel('Ship Mode',fontsize=12)
plt.legend(loc='upper center');
```

```python
fig,ax=plt.subplots(figsize=(9,5))
sns.barplot(x='Category',y='Sales',hue='Segment',data=df,ci=False)
plt.ylabel('Sales',fontsize=12)
plt.xlabel('Category',fontsize=12)
plt.legend(loc='upper center');
```

```python
fig,ax=plt.subplots(figsize=(9,5))
sns.barplot(x='Category',y='Sales',hue='Region',data=df,ci=False)
plt.ylabel('Sales',fontsize=15)
plt.xlabel('Category',fontsize=15)
plt.legend(loc='upper center');
```

```python
fig,ax=plt.subplots(figsize=(9,5))
df.groupby('Sub-Category')['Sales'].agg(np.mean).sort_values().plot(kind='barh',color='#ab4242',ylabel='Sub-Category',fontsize=12)
plt.ylabel('Sub-Categories',fontsize=14)
plt.xlabel('Sales AVG',fontsize=14);
```
